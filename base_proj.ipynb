````python
import numpy as np
import cv2

## Criando uma função para transformar a imagem e normalizar elas
def normaliza_image(images_list):
  images = []
  for img in images_list:
    img_array = np.array(img)
    image = cv2.resize(img_array, (75, 75), 0, 0, cv2.INTER_LINEAR)
    images.append(image)
  return np.array(images)

## Preparando os dados

# train_data_preparado = (train_data['labels'], normaliza_image(train_data['image']))
# validation_data_preparado = (train_data['labels'], normaliza_image(train_data['image']))
# test_data_preparado = (train_data['labels'], normaliza_image(train_data['image']))
````

````python
from tensorflow.keras.utils import to_categorical

train_newImages = normaliza_image(train_data['image'])
train_labelsNp = np.array(train_data['labels'])

test_newImages = normaliza_image(test_data['image'])
test_labelsNp = np.array(test_data['labels'])

validation_newImages = normaliza_image(validation_data['image'])
validation_labelsNp = np.array(validation_data['labels'])

train_labels_one_hot = to_categorical(train_labelsNp)
test_labels_one_hot = to_categorical(test_labelsNp)
validation_labels_one_hot = to_categorical(validation_labelsNp)
````

````python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model

# Carregar o modelo pré-treinado ResNet50 sem as camadas densas (fully connected)
resnet50 = ResNet50(include_top=False, weights='imagenet', input_shape=(75, 75, 3))

# Congelar os pesos das camadas convolucionais existentes
for layer in resnet50.layers:
    layer.trainable = False

# Adicionar camadas densas personalizadas no topo da ResNet50
x = Flatten()(resnet50.output)
x = Dense(256, activation='relu')(x)
output = Dense(len(train_labels_one_hot[0]), activation='softmax')(x)

# Criar o modelo final
model_resnet50 = Model(inputs=resnet50.input, outputs=output)
````

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Criação do gerador de aumento de dados
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
````

````python
from sklearn.model_selection import train_test_split

train_images, val_images, train_labels, val_labels = train_test_split(train_newImages, train_labels_one_hot, test_size=0.1, random_state=42)
````

````python
# specify the loss function and optimizer
# model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

model_resnet50.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# train the data for 500 epochs
history = model_resnet50.fit(datagen.flow(train_images, train_labels, batch_size=32), epochs=100, verbose=1, validation_data=(val_images, val_labels))
# save the model
model_resnet50.save('pokedex.model')
````

````python
model_resnet50.save('modelo1.h5')
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Suponha que 'y_true' são os rótulos verdadeiros e 'y_pred' são as previsões do modelo
predicoes_images = model_resnet50.predict(test_newImages)
# Acurácia
# print(predicoes_images[321])
# print(len(test_labels_one_hot))

predicted_classes = np.argmax(predicoes_images, axis=1)
predicted_classes2 = np.argmax(test_labels_one_hot, axis=1)

accuracy = accuracy_score(predicted_classes2, predicted_classes)

# Precisão
precision = precision_score(predicted_classes2, predicted_classes, average='weighted')

# Recall
recall = recall_score(predicted_classes2, predicted_classes, average='weighted')

# F1-score
f1 = f1_score(predicted_classes2, predicted_classes, average='weighted')

print("Acurácia:", accuracy)
print("Precisão:", precision)
print("Recall:", recall)
print("F1-score:", f1)
````
